```python
import re

def simple_chatbot(user_input):

    user_input = user_input.lower()


    patterns = {
        'greeting': r'^(hi|hello|hey|good morning|good afternoon)\b',
        'age_question': r'^(how old are you|what is your age|age)\b',
        'name_question': r'^(what is your name|who are you)\b',
        'weather_question': r'^(what\'s the weather like today|how\'s the weather)\b',
        'help_request': r'^(help|can you help me|assist me)\b',
        'gratitude': r'^(thank you|thanks)\b',
        'goodbye': r'^(bye|goodbye|see you later|take care)\b'
    }

    responses = {
        'greeting': "Hello! How can I assist you today?",
        'age_question': "I am just a computer program and don't have an age!",
        'name_question': "I am a chatbot created to assist you.",
        'weather_question': "I'm sorry, I don't have access to real-time weather information.",
        'help_request': "Sure, I'll do my best to help you. What do you need assistance with?",
        'gratitude': "You're welcome!",
        'goodbye': "Goodbye! Have a great day."
    }


    for intent, pattern in patterns.items():
        if re.search(pattern, user_input):
            return responses[intent]


    return "Sorry, I didn't understand that. Can you please ask again?"


print("Welcome! Ask me anything or say goodbye to exit.")
while True:
    user_input = input("You: ")
    if user_input.lower() in ['exit', 'quit', 'goodbye', 'bye']:
        print("Bot:", simple_chatbot(user_input))
        break
    else:
        print("Bot:", simple_chatbot(user_input))
```
